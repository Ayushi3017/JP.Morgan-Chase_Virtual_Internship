```python
import pandas as pd
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("nat_gas.csv")
```

```python
df.columns
```

```python
df['Dates'] = pd.to_datetime(df['Dates'])
```

```python
df['Dates'] = pd.to_datetime(df['Dates'])

# Function to estimate price
def estimate_price(target_date):
    target_date = pd.to_datetime(target_date)
    prices_series = df.set_index('Dates')['Prices']  # use exact column name
    return np.interp(
        target_date.value,
        prices_series.index.values.astype(np.int64),
        prices_series.values
    )

# Injection simulation parameters
injection_dates = pd.to_datetime(['2025-01-15', '2025-02-15', '2025-03-15'])
injection_rate = 10
max_storage = 50
stored_volume = 0

# Loop through injection dates
for date in injection_dates:
    price = estimate_price(date)
    
    # Allowed volume to inject
    volume = min(injection_rate, max_storage - stored_volume)
    
    # Update stored volume
    stored_volume += volume
    
    # Cost of injection
    cost = volume * price
    
    print(f"Date: {date.date()}, Price: {price:.2f}, Injected: {volume}, Stored Volume: {stored_volume}, Cost: {cost:.2f}")
```
